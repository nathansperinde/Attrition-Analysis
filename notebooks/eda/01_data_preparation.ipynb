```python
df_employee_attrition["AttritionFlag"] = df_employee_attrition["Attrition"].map({
        "Yes": 1,
        "No": 0
    })

attrition_rate = df_employee_attrition["AttritionFlag"].mean()

print(f'Taxa de Turnover: {attrition_rate:.2%}')
```

```python
df_employee_attrition["AgeGroup"] = pd.cut(
    df_employee_attrition["Age"],
    bins=[17, 25, 35, 45, 55, 65],
    labels=["18-25", "26-35", "36-45", "46-55", "56-65"]
)

df_employee_attrition["TenureGroup"] = pd.cut(
    df_employee_attrition["YearsAtCompany"],
    bins=[-1, 1, 3, 5, 10, 40],
    labels=["0-1", "2-3", "4-5", "6-10", "10+"]
)

df_employee_attrition["DistanceGroup"] = pd.cut(
    df_employee_attrition["DistanceFromHome"],
    bins=[0, 5, 10, 20, 30],
    labels=["0-5", "6-10", "11-20", "21-30"],
    include_lowest=True
)

df_employee_attrition["CareerStage"] = pd.cut(
    df_employee_attrition["TotalWorkingYears"],
    bins=[-1, 2, 5, 10, 20, 40],
    labels=["Entry-level", "Early career", "Mid career", "Experienced", "Late career"]
)

df_employee_attrition["RoleStabilityGroup"] = pd.cut(
    df_employee_attrition["YearsInCurrentRole"],
    bins=[-1, 1, 3, 7, 20],
    labels=["0-1", "2-3", "4-7", "8+"]
)

df_employee_attrition["PromotionDelayGroup"] = pd.cut(
    df_employee_attrition["YearsSinceLastPromotion"],
    bins=[-1, 0, 2, 5, 15],
    labels=["Promoted this year", "1-2 years", "3-5 years", "5+ years"]
)

df_employee_attrition["IncomeGroup"] = pd.cut(
    df_employee_attrition["MonthlyIncome"],
    bins=[0, 3000, 6000, 10000, 20000],
    labels=["Low", "Medium", "High", "Very High"]
)


education_map = {
    1: "Below College",
    2: "College",
    3: "Bachelor",
    4: "Master",
    5: "Doctor"
}

satisfaction_map = {
    1: "Low",
    2: "Medium",
    3: "High",
    4: "Very High"
}

performance_rating_map = {
    1: "Low",
    2: "Good",
    3: "Excellent",
    4: "Outstanding"
}

work_life_balance_map = {
    1: "Bad",
    2: "Good",
    3: "Better",
    4: "Best"
}


df_employee_attrition["EducationLevel"] = (
    df_employee_attrition["Education"].map(education_map)
)

df_employee_attrition["StockOption"] = df_employee_attrition["StockOptionLevel"].map(
    {
        0: "No Options",
        1: "Low",
        2: "Medium",
        3: "High"
    }
)

df_employee_attrition["JobLevelGroup"] = df_employee_attrition["JobLevel"].map(
    {
        1: "Entry Level",
        2: "Junior Level",
        3: "Mid Level",
        4: "Senior Level",
        5: "Executive Level"
    }
)

df_employee_attrition["SatisfactionLevel"] = (
    df_employee_attrition["JobSatisfaction"].map(satisfaction_map)
)

df_employee_attrition["PerformanceRatingLevel"] = (
    df_employee_attrition["PerformanceRating"].map(performance_rating_map)
)

df_employee_attrition["EnvironmentSatisfactionLevel"] = (
    df_employee_attrition["EnvironmentSatisfaction"].map(satisfaction_map)
)

df_employee_attrition["JobInvolvementLevel"] = (
    df_employee_attrition["JobInvolvement"].map(satisfaction_map)
)

df_employee_attrition["RelationshipSatisfactionLevel"] = (
    df_employee_attrition["RelationshipSatisfaction"].map(satisfaction_map)
)

df_employee_attrition["WorkLifeBalanceLevel"] = (
    df_employee_attrition["WorkLifeBalance"].map(work_life_balance_map)
)
```

```python
mapped_columns = [
    "EducationLevel",
    "StockOption",
    "JobLevelGroup",
    "SatisfactionLevel",
    "PerformanceRatingLevel",
    "EnvironmentSatisfactionLevel",
    "JobInvolvementLevel",
    "RelationshipSatisfactionLevel",
    "WorkLifeBalanceLevel"
]

df_employee_attrition[mapped_columns].isna().sum()
```
